# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
dr = df[df['Medal'] == 'Gold']
data = []

for Sport in dr['Sport'].unique():
    data.append([Sport , len(dr[dr['Sport']  == Sport])])

pd.DataFrame(data,columns = ['Sport','Gold']).sort_values(by='Gold', ascending=False).head(5).plot(x = 'Sport', y = 'Gold', kind = 'bar', figsize = (5,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []

for Sport in df['Sport'].unique():
    data.append([Sport , len(df[df['Sport']  == Sport])])

pd.DataFrame(data,columns = ['Sport','Medals']).sort_values(by='Medals', ascending=False).head(5).plot(x = 'Sport', y = 'Medals', kind = 'bar', figsize = (5,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []

for Athlete in df['Athlete'].unique():
    data.append([Athlete , len(df[df['Athlete']  == Athlete])])

pd.DataFrame(data,columns = ['Athlete','Playmed']).sort_values(by='Playmed', ascending=False).head(5).plot(x = 'Athlete', y = 'Playmed', kind = 'bar', figsize = (5,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
dr = df[df['Medal'] == 'Gold']
data = []

for Athlete in dr['Athlete'].unique():
    data.append([Athlete , len(dr[dr['Athlete']  == Athlete])])

pd.DataFrame(data,columns = ['Athlete','Playgmed']).sort_values(by='Playgmed', ascending=False).head(5).plot(x = 'Athlete', y = 'Playgmed', kind = 'bar', figsize = (5,5))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for Event in df['Event'].unique():
    data.append([Event , len(df[df['Event']  == Event])])

pd.DataFrame(data,columns = ['Event','Players']).sort_values(by='Players', ascending=False).head().plot(x = 'Event', y = 'Players', kind = 'bar', figsize = (5,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
dr = df[df['Medal'] == 'Gold']
dg = dr[dr['Gender'] == 'Women']
data = []

for Sport in dg['Sport'].unique():
    data.append([Sport , len(dg[dg['Sport']  == Sport])])

pd.DataFrame(data,columns = ['Sport','Women']).sort_values(by='Women', ascending=False).head(5).plot(x = 'Sport', y = 'Women', kind = 'bar', figsize = (5,5))
```
